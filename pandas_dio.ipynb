```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

# Creazione dati artificiali
N = 100
r0 = 0.6
x = 0.9 * np.random.rand(N)
y = 0.9 * np.random.rand(N)
area = (20 * np.random.rand(N))**2  # 0 to 10 point radii


# Impostazioni grafico
c = np.sqrt(area)
r = np.sqrt(x ** 2 + y ** 2)
area1 = np.ma.masked_where(r < r0, area)
area2 = np.ma.masked_where(r >= r0, area)
plt.scatter(x, y, s=area1, marker='^', c=c)
plt.scatter(x, y, s=area2, marker='o', c=c)
# Show the boundary between the regions:
theta = np.arange(0, np.pi / 2, 0.01)
plt.plot(r0 * np.cos(theta), r0 * np.sin(theta))

plt.show()
```

```python
# Generare dati artificiali distribuiti normalmente
# np.random.seed(42)  # Imposta il seed per riproducibilità dei risultati
data = np.random.normal(
    loc=0,    # Media della distribuzione (centro dei dati)
    scale=1,  # Deviazione standard (dispersione dei dati)
    size=1000 # Numero di campioni da generare
)

# Calcolare i quartili
q1, q2, q3 = np.percentile(
    data,       # Array dei dati
    [25, 50, 75] # Percentili da calcolare (25% = Q1, 50% = mediana, 75% = Q3)
)

# Creare l'istogramma
plt.figure(
    figsize=(10, 5), # Dimensioni della figura (larghezza, altezza in pollici)
    facecolor="b" # Colore di sfondo della figura
)

# Cambiare solo lo sfondo dell'area del grafico (dietro le colonne)
ax = plt.gca()  # Ottieni l'oggetto Axes corrente
ax.set_facecolor('black')  # Imposta lo sfondo dell'area del grafico a nero

# Modificare i colori degli assi e dei testi per contrasto
ax.spines['bottom'].set_color('white')  # Colore asse X
ax.spines['left'].set_color('white')    # Colore asse Y
ax.tick_params(axis='x', colors='white') # Colore numeri asse X
ax.tick_params(axis='y', colors='white') # Colore numeri asse Y

plt.hist(
    data, 
    bins=70,          # Numero di contenitori/barre nell'istogramma (può essere anche una lista o una stringa per algoritmi automatici)
    color="skyblue",  # Colore di riempimento delle barre
    edgecolor="black", # Colore dei bordi delle barre
    alpha=0.7,        # Trasparenza (0 = trasparente, 1 = opaco)
    density=False,    # Se True, normalizza l'area a 1 (probabilità invece che conteggi)
    # cumulative=True, # Mostra la distribuzione cumulativa
    # label='Dati',    # Etichetta per la legenda
    # orientation='horizontal', # Orientamento dell'istogramma
    # histtype='step', # Tipo di istogramma ('bar', 'barstacked', 'step', 'stepfilled')
    # log=True,        # Scala logaritmica sull'asse y
    # stacked=True,    # Sovrappone più istogrammi
    # align='mid',     # Allineamento barre ('left', 'mid', 'right')
    # range=(-3, 3)    # Intervallo valori da considerare
)

# Aggiungere linee VERTICALI (AXVLINE) per i quartili
plt.axvline( # Orizzontali con AXHLINE
    q1, 
    color="red",        # Colore della linea
    linestyle="dashed", # Stile della linea (':', '--', '-.', etc.)
    linewidth=2,        # Spessore della linea
    label="1° Quartile (Q1)" # Etichetta per legenda
)
plt.axvline(q2, color="green", linestyle="dashed", linewidth=2, label="Mediana (Q2)")
plt.axvline(q3, color="purple", linestyle="dashed", linewidth=2, label="3° Quartile (Q3)")

# Aggiungere etichette e titolo
plt.xlabel("Valore")          # Etichetta asse X
plt.ylabel("Frequenza")       # Etichetta asse Y
plt.title("Istogramma con Quartili") # Titolo del grafico
plt.legend(                   # Mostra la legenda
    loc='best',               # Posizione automatica della legenda
    frameon=True,             # Mostra lo sfondo
    shadow=True               # Aggiungi ombra
)
plt.grid(
    axis="y",       # Mostra griglia solo per l'asse Y ('both', 'x', 'y')
    linestyle="--", # Stile delle linee della griglia
    alpha=0.7       # Trasparenza della griglia
)

# Personalizzazione aggiuntiva
plt.xlim(-4, 4)      # Limiti dell'asse X
# plt.ylim(0, 100)   # Limiti dell'asse Y
plt.tight_layout()   # Migliora il layout automaticamente

# Mostrare il grafico
plt.show()
```

```python
balance_not_student = pippo[pippo.Student == "No"]["Balance"]
balance_student = pippo[pippo.Student == "Yes"]["Balance"]

max_val = max(balance_not_student.max(), balance_student.max())
min_val = min(balance_not_student.min(), balance_student.min())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15,5),)
bins_edge = np.linspace(min_val, max_val, 11)

fig.suptitle("Grafico Doppio")

sns.histplot(
    balance_not_student,
    ax=axes[0],
    kde=True,
    color="dodgerblue",
    bins=bins_edge,
)

sns.histplot(
    balance_student,
    ax=axes[1],
    kde=True,
    color="darkred",
    bins=bins_edge,
)

axes[0].set_title("Not students")
axes[1].set_title("Students")

plt.tight_layout()
```

```python
#Grafico in 4: studenti sposati vs non sposati e non stud sposati vs non stud non spos
fig, axes = plt.subplots(2, 2, figsize=(10,10))
fig.suptitle("Grafico quadruplo")
n_bins = [100,200,300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500,1600,1700,1800,1900,2000]

# Studenti sposati
sns.histplot(
    pippo[pippo.Student == "Yes"][pippo.Married == "Yes"]["Balance"],
    ax=axes[0,0],
    kde=True,
    color="dodgerblue",
    bins=n_bins    
    )
axes[0,0].set_title("Studenti sposati")
axes[0,0].set_xlim(0,2000)
axes[0,0].set_ylim(0,20)

# Studenti non sposati
sns.histplot(
    pippo[pippo.Student == "Yes"][pippo.Married == "No"]["Balance"],
    ax=axes[0,1],
    kde=True,
    color="lightsteelblue",
    bins=n_bins
)
axes[0,1].set_title("Studenti non sposati")
axes[0,1].set_xlim(0,2000)
axes[0,1].set_ylim(0,20)

# Non studenti sposati
sns.histplot(
    pippo[pippo.Student == "No"][pippo.Married == "Yes"]["Balance"],
    ax=axes[1,0],
    kde=True,
    color="gold",
    bins=n_bins
)
axes[1,0].set_title("Non studenti sposati")
axes[1,0].set_xlim(0,2000)
axes[1,0].set_ylim(0,20)

# Non studenti non sposati
sns.histplot(
    pippo[pippo.Student == "No"][pippo.Married == "No"]["Balance"],
    ax=axes[1,1],
    kde=True,
    color="goldenrod",
    bins=n_bins
)
axes[1,1].set_title("Non studenti non sposati")
axes[1,1].set_xlim(0,2000)
axes[1,1].set_ylim(0,20)

plt.tight_layout()
```
